```python
import wandb
import pandas as pd

pd.options.display.max_columns = None
```

```python
# 1. Initialize API and fetch runs
api = wandb.Api()
runs = api.runs("c5-team2/yolo-eval") # Replace with your entity and project
```

```python
# 2. Extract metrics
summary_list = []
summary_metrics_list = []
history_list = []

for run in runs:
    # Get metadata/summary
    summary_list.append(run.summary._json_dict)
    
    # Get config (hyperparameters)
    summary_metrics_list.append(run.config)
    
    # Get full history (step-by-step metrics)
    history_df = pd.DataFrame(run.scan_history())
    history_df['run_name'] = run.name
    history_list.append(history_df)
```

```python
# 3. Save as CSV
# Save summary/config (one row per run)
summary_df = pd.DataFrame(summary_list)
summary_df.to_csv("../../results/task_d/yolo_eval.csv", index=False)
```
